```python
%%js
class FrogSpecies {
    constructor(name, size, habitat, predators, diet, lifespan, toxicity, funFacts) {
        this.name = name;
        this.size = size;
        this.habitat = habitat;
        this.predators = predators;
        this.diet = diet;
        this.lifespan = lifespan;
        this.toxicity = toxicity;
        this.funFacts = funFacts;
    }

    print() {
        console.log(`Species: ${this.name}`);
        console.log(`Size: ${this.size}`);
        console.log(`Habitat: ${this.habitat}`);
        console.log(`Predators: ${this.predators}`);
        console.log(`Diet: ${this.diet}`);
        console.log(`Lifespan: ${this.lifespan}`);
        console.log(`Toxicity: ${this.toxicity}`);
        console.log(`Fun Facts: ${this.funFacts}`);
        console.log('----------------------------------------');
    }
}

// Create an array to store frog instances
var frogSpeciesArray = [];

// Function to add a new frog species to the array
function addFrogSpecies(name, size, habitat, predators, diet, lifespan, toxicity, funFacts) {
    var frog = new FrogSpecies(name, size, habitat, predators, diet, lifespan, toxicity, funFacts);
    frogSpeciesArray.push(frog);
}

// Add frog species to the array
addFrogSpecies(
    "American Bullfrog (Lithobates catesbeianus),"
    "Small",
    "Central America",
    "Snakes, birds",
    "Insects",
    "5 years",
    "Not Poisonous",
    "They sleep with their eyes closed."
);

addFrogSpecies(
    "Dart Frog",
    "Tiny",
    "Central and South America",
    "Snakes, birds",
    "Ants, mites",
    "3 years",
    "Poisonous",
    "Some species are used by indigenous people to poison their blowgun darts."
);

// Add more frog species to the array...

// Loop through the array and print details for each frog
for (var frog of frogSpeciesArray) {
    frog.print();
}
```
