```python
print("Hello World!")
msg = input("Enter a greeting: ")
print(msg)
def question_and_answer(prompt):
    print("Question: " + prompt)
    msg = input()
    print("Answer: " + msg)

question_and_answer("Are you ready for a frog quiz?")
question_and_answer("Whats your favorite frog?")
question_and_answer("The quiz will be on the Sumaco Horned Frog and will be 3 questions, do you know that frog?")
import getpass, sys

def question_with_response(prompt):
    print("Question: " + prompt)
    msg = input()
    return msg

questions = 3
correct = 0

print('Hello, ' + getpass.getuser() + " running " + sys.executable)
print("You will be asked " + str(questions) + " questions.")
question_and_answer("Are you ready to take a test?")

rsp = question_with_response("Where does the Sumaco Horned Frog live?")
if rsp == "Ecuador":
    print(rsp + " is correct!")
    correct += 1
else:
    print(rsp + " is incorrect!")

rsp = question_with_response("What does the Sumaco Horned Frog eat?")
if rsp == "insects and small invertebrates":
    print(rsp + " is correct!")
    correct += 1
else:
    print(rsp + " is incorrect!")

rsp = question_with_response("How long does the Sumaco Horned Frog live?")
if rsp == "8 to 10 years":
    print(rsp + " is correct!")
    correct += 1
else:
    print(rsp + " is incorrect!")

print(getpass.getuser() + " you scored " + str(correct) +"/" + str(questions))
```
